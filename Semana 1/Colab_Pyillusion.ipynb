# **Reprodução das bandas de Mach (adaptação ao brilho)**

```python
# Importar bibliotecas
import cv2 # biblioteca OpenCV
import numpy as np # biblioteca "Numerical Python"
from google.colab.patches import cv2_imshow # biblioteca para apresentação de imagens cv2 em Colab

# Dimensões da imagem
largura = 800
altura = 600

# Criar uma matriz de pontos escuros (zeros) com as dimensões acima
imagem = np.zeros((altura, largura),dtype=np.uint8)

# Número de barras verticais
num_barras = 8

# Largura de cada barra
largura_barra = largura // num_barras

# Loop para criar as barras
for i in range(num_barras):
   # Calcular os limites horizontais da barra
   inicio = i * largura_barra
   fim = (i + 1) * largura_barra
   # Preencher a barra com tons de cinza, variando de
   # acordo com a posição
   valor_tom_cinza = int(255 * i / (num_barras - 1))
   imagem[:, inicio:fim] = valor_tom_cinza

# Mostrar a imagem gerada
cv2_imshow(imagem)

# Salvar a imagem em disco
Res = cv2.imwrite('barras_verticais.png', imagem)


```

# **Ilusão de ótica gerada por programação em Python**

```python
from PIL import Image
import cv2
from google.colab.patches import cv2_imshow
import requests

def crop_square(img, size, interpolation=cv2.INTER_AREA):
    h, w = img.shape[:2]
    min_size = np.amin([h,w])

    # Centralize and crop
    crop_img = img[int(h/2-min_size/2):int(h/2+min_size/2), int(w/2-min_size/2):int(w/2+min_size/2)]
    resized = cv2.resize(crop_img, (size, size), interpolation=interpolation)

    return resized

print("Exhibition of artists and robots at the Paris Grand Palais - Pedro Szekely\n\n")

headers = {
    'User-Agent': 'My User Agent 1.0'
}

picture_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/e/e1/Artists_and_Robots_at_the_Paris_Grand_Palais_%2842158514392%29.jpg/1920px-Artists_and_Robots_at_the_Paris_Grand_Palais_%2842158514392%29.jpg"
# picture_url = "https://upload.wikimedia.org/wikipedia/commons/
#  wikimedia.org/wiki/Optical_illusion#/media/File:Artists_and_Robots_at_the_Paris_Grand_Palais_(42158514392).jpg"
# picture_url = "https://upload.wikimedia.org/wikipedia/commons/d/d6/Aile_de_Tachinidae.png?20210222185924"
# picture_url = "https://upload.wikimedia.org/wikipedia/commons/6/63/Arrecifes_de_Cozumel_National_Park.jpg"
# picture_url = "https://upload.wikimedia.org/wikipedia/commons.wikimedia.org/wiki/File:St._Clears-_The_Town_Hall_(geograph_6883892).jpg"
# picture_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/e/ea/Van_Gogh_-_Starry_Night_-_Google_Art_Project.jpg/2728px-Van_Gogh_-_Starry_Night_-_Google_Art_Project.jpg"
r = requests.get(picture_url, headers=headers, stream=True)
# Image.open(r.raw)
# r = requests.get(picture_url, headers=headers, stream=True)
im1 = Image.open(r.raw)
im1 = im1.save("optic.jpg")

img = cv2.imread("optic.jpg")
img = crop_square(img, 800)
# # Downsize without aspect ratio
# img = cv2.resize(img, (500,500), interpolation=cv2.INTER_AREA)
cv2_imshow(img)
```

# **Ilusão de Delboeuf**
A ilusão de Delboeuf é uma ilusão de ótica de percepção de tamanho relativo, onde círculos de tamanho idêntico parecem diferentes devido ao contexto circundante. A ilusão recebeu o nome do filósofo, matemático, psicólogo experimental, hipnotizador e psicofísico belga Joseph Remi Leopold Delboeuf (1831-1896), que a criou em 1865.

```python
import pyllusion

delboeuf = pyllusion.Delboeuf(illusion_strength=3)
delboeuf.to_image()
```

# **Ilusão de Ebbinghaus**
A ilusão de Ebbinghaus é uma ilusão de ótica de percepção de tamanho relativo, onde círculos de tamanho idêntico parecem diferentes devido ao contexto circundante (o círculo vermelho direito parece menor porque o círculo circundante é maior). A ilusão recebeu o nome de seu criador, o psicólogo alemão Hermann Ebbinghaus (1850–1909), embora tenha sido popularizada por Edward B. Titchener em um livro de psicologia experimental de 1901.

```python
ebbinghaus = pyllusion.Ebbinghaus(illusion_strength=2)
ebbinghaus.to_image()
```

# **Ilusão de Müller-Lyer**
A ilusão de Müller-Lyer é uma ilusão de ótica que faz com que o participante perceba dois segmentos como sendo de comprimentos diferentes dependendo do formato das setas. A ilusão recebeu o nome de seu criador, o sociólogo alemão Franz Carl Müller-Lyer (1857–1916) em 1889.



```python
mullerlyer = pyllusion.MullerLyer(illusion_strength=30)
mullerlyer.to_image()
```

# **Ilusão Ponzo**
A ilusão Ponzo é uma ilusão de ótica de percepção de tamanho relativo, onde linhas horizontais de tamanho idêntico aparecem como diferentes devido ao contexto circundante (a linha superior aparece mais longa, pois é interpretada como estando à distância). A ilusão recebeu o nome de seu criador, o psicólogo italiano Mario Ponzo (1882–1960) em 1911, que sugeriu que a mente humana julga o tamanho de um objeto com base em seu fundo.

```python
ponzo = pyllusion.Ponzo(illusion_strength=20)
ponzo.to_image()
```

# **Ilusão de Zöllner**
A ilusão de Zöllner é uma ilusão de ótica, onde as linhas horizontais são percebidas como não paralelas devido ao seu fundo. É assim nomeada após seu descobridor, o astrofísico alemão Johann Karl Friedrich Zöllner em 1860.

```python
zollner = pyllusion.Zollner(illusion_strength=75)
zollner.to_image()
```
